```python
import pandas as pd
import requests
import json
```

```python
from config import api_key
```

```python
url = "https://api.rawg.io/api/games?"
page = 1
```

```python
query_url = f"{url}key={api_key}&page:{page}"
```

```python
name_list = []
metacritic_list = []
release_list = []
genre_list = []

for page in range(1, 400):
    query_url = f"{url}key={api_key}&page={page}"
    response = requests.get(query_url)
    data = response.json()
    for game in data["results"]:
        name = game["name"]
        name_list.append(name)
        metacritic_rating = game["metacritic"]
        metacritic_list.append(metacritic_rating)
        released_date = game["released"]
        release_list.append(released_date)
        genre = game["genres"][0]["name"] if game["genres"] else None
        genre_list.append(genre)
 
```

```python
game_api_df = pd.DataFrame(
    {"Name": name_list,
     "Metacritic Rating":metacritic_list,
     "Release Date":release_list,
     "Genre":genre_list,    
    })
game_api_df
```

```python
game_api_df.to_csv('game_api.csv')
```
