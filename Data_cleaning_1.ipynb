```python
from urllib.parse import quote_plus
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
#from config import pwd_postgresql

```

```python
#videogame_sales data
csv_file = "data/vgsales.csv"
sales_df = pd.read_csv(csv_file)
sales_df.head()
```

```python
new_sales_df = sales_df[['Name', 'Platform','Year','Publisher','Developer','Global_Sales']].copy()
new_sales_df.head()
```

```python
#new_sales_df['id']=new_sales_df.index
#new_sales_df
```

```python
#new_sales_df_set = new_sales_df.set_index('id')
#new_sales_df_set
```

```python
#replacing NaNs
salesn_df =new_sales_df.replace(np.nan, 0)
salesn_df.head()
```

```python
salesn_df["Year"]=salesn_df["Year"].astype('int64')
salesn_df.head()
```

```python
#Rename/Organize Columns
renamed_salesn_df =salesn_df.rename(columns={"Name":"name","Platform":"platform","Year":"year","Publisher":"publishers","Developer":"developer","Global_Sales":"global_sales"})
renamed_salesn_df
```
